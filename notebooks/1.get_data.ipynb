```python
start_date = '2000-01-01'
end_date = '2026-01-15'
```

```python
import yfinance as yf

gold = yf.download("GC=F", start=start_date, end=end_date, interval="1d")
gold = gold[['Close']].reset_index()
gold.columns = ['date', 'price']

silver = yf.download("SI=F", start=start_date, end=end_date, interval="1d")
silver = silver[['Close']].reset_index()
silver.columns = ['date', 'price']
```

```python
gold.to_csv('data/daily/gold.csv', index=False)
silver.to_csv('data/daily/silver.csv', index=False)
```
